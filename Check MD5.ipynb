
# Check MD5



### import


```python
import os
import glob
import hashlib
```


### function def

Expected file names : file_to_check file_to_check.md5  
With the checksum on the 1st line

```python
def check_checksum(file_to_check):
    ### compute MD5 checksum for file_to_check
    md5_hash = hashlib.md5()
    infile = open(file_to_check, "rb")
    content = infile.read()
    md5_hash.update(content)
    digest = md5_hash.hexdigest()
    infile.close()
    print("MD5 checksum : \n" + digest)

    ### get checksum in MD5 file
    file_md5 = file_to_check + '.md5'
    with open(file_md5, 'r', encoding='utf-8') as infile_md5:
        checksum = infile_md5.readline().split()[0]
        print("MD5 checksum in verification file : \n" + checksum)

    ### compare computed checksum and verification file's checksum
    if digest == checksum :
        print("OK ; both checksum are the same")
        return 0
    else :
        print("/!\\/!\\/!\\ NOT OK ; Checksums are different : check your file /!\\/!\\/!\\")
        return 1
```


## For 1 file



### var def


```python
# (Windows) Don't forget to take the string __raw__, as in windows, paths use backslashes
file_to_check = r"C:\Users\vguerlais\Desktop\RNAseq\30-536685438\00_fastq\1_R1_001.fastq.gz"
```


### exec


```python
check_checksum(file_to_check)
```


## For a dir



### var def


```python
# Don't forget to take the string __raw__, as in windows, path uses backslashes
dir_to_check = r"C:\Users\vguerlais\Desktop\RNAseq\30-536685438\00_fastq"
```


## exec


```python
os.chdir(dir_to_check)

# to get a check on the whole dir
all_files_ok = True

# I'm gathering md5 file to ignore unconcered files in the dir
list_of_files = glob.glob("*.md5")
for md5_files in list_of_files :
    file_to_check = md5_files.rstrip('.md5')
    print("Checking file : " + file_to_check)
    file_ok = check_checksum(file_to_check)
    if file_ok == 1 :
        all_files_ok = False
    print("\n")

# printing a result for the dir
if all_files_ok :
    print("All files checked, everything's good !")
else :
    print("ERROR ! At least 1 file's checksum does not match. Check log above")
```
